```python
# Step 2: Upload PDF and extract text
from google.colab import files
import fitz  # PyMuPDF

uploaded = files.upload()
pdf_text = ""

for file_name in uploaded:
    with fitz.open(file_name) as doc:
        for page in doc:
            pdf_text += page.get_text()

print("✅ PDF uploaded and text extracted. Preview:")
print(pdf_text[:1000])  # show first 1000 characters
```

```python
# Step 3: Setup Google Generative AI and list available models
import google.generativeai as genai

# Use the GOOGLE_API_KEY from the secrets manager
from google.colab import userdata
genai.configure(api_key=API_KEY)

print("🔍 Fetching available models for your API key...")
available_models = genai.list_models()

# Pick a preferred model if available
preferred_models = [
    "models/gemini-1.5-flash",
    "models/gemini-pro",
    "models/chat-bison-001",
    "models/text-bison-001",
]

model_name = None
for model in available_models:
  if model.name in preferred_models:
    model_name = model.name
    break

if model_name is None:
    raise ValueError("No supported models available for your API key.")

print(f"✅ Using model: {model_name}")
model = genai.GenerativeModel(model_name)
```

```python
# Step 4: Create UI for asking questions about the PDF content
from IPython.display import display
import ipywidgets as widgets

query_input = widgets.Text(
    value='',
    placeholder='Ask a question about the PDF...',
    description='Question:',
    layout=widgets.Layout(width='80%')
)

ask_button = widgets.Button(description="Ask AI", button_style='success')
output_area = widgets.Output()

def on_click(b):
    with output_area:
        output_area.clear_output()
        query = query_input.value.strip()
        if not query:
            print("⚠️ Please enter a question.")
            return

        prompt = f"""
You are an expert assistant helping summarize and interpret documents.
Use the following document content to answer the question.

Document content:
{pdf_text[:15000]}

Question:
{query}
"""
        response = model.generate_content(prompt)
        print("💬 Answer:\n")
        print(response.text)

ask_button.on_click(on_click)
display(query_input, ask_button, output_area)
```
